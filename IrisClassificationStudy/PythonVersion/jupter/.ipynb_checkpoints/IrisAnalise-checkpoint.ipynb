```python
iris_df["Names"] = iris_df["Names"].map(name_x_cat)
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = iris_df_std
Y = iris_df.iloc[:, -1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=45, stratify=Y)
```

# Perseptron

```python
from sklearn.linear_model import Perceptron
```

# Logistic Regression CV and Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
```

```python
X_train_bin = X_train[(Y_train == 0) | (Y_train == 1)]
Y_train_bin = Y_train[(Y_train == 0) | (Y_train == 1)]

X_test_bin = X_test[(Y_test == 0) | (Y_test == 1)]
Y_test_bin = Y_test[(Y_test == 0) | (Y_test == 1)]

lrCV = LogisticRegressionCV()
lrCV.fit(X_train_bin, Y_train_bin)

y_pred_lrCV = lrCV.predict(X_test_bin)

lrCV_percent = (y_pred_lrCV == Y_test_bin).sum() / len(Y_test_bin)
lrCV_acc = accuracy_score(y_pred_lrCV, Y_test_bin)
lrCV_acc
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)

y_pred_lr = lr.predict(X_test)

lr_percent = (y_pred_lr == Y_test).sum() / len(Y_test)
lr_acc = accuracy_score(y_pred_lr, Y_test)
lr_acc
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)

y_pred_dt = dt.predict(X_test)
dt_percent = (y_pred_dt == Y_test).sum() / len(Y_test)
dt_acc = accuracy_score(y_pred_dt, Y_test)
dt_acc
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)

y_pred_rf = rf.predict(X_test)
rf_perdent = (y_pred_rf == Y_test).sum() / len(Y_test)
rf_acc = accuracy_score(y_pred_rf, Y_test)
rf_acc
```

# Support Vector Machines

```python
from sklearn.svm import SVC

svm = SVC(kernel="linear")
svm.fit(X_train, Y_train)

y_pred_svm = svm.predict(X_test)

svm_percent = (y_pred_svm == Y_test).sum() / len(Y_test)
svm_acc = accuracy_score(y_pred_svm, Y_test)
svm_acc
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)

y_pred_knn = knn.predict(X_test)
knn_percent = (y_pred_knn == Y_test).sum() / len(Y_test)
knn_acc = accuracy_score(y_pred_knn, Y_test)
knn_acc
```

```python
acc_result_list = [pc_acc, lrCV_acc, lr_acc, dt_acc, rf_acc, svm_acc, knn_acc]
df_result = pd.DataFrame(acc_result_list).T
df_result.columns = ["PC", "LRCV(BIN)", "LR", "DT", "RF", "SVM", "KNN"]
sns.catplot(data=df_result, kind="bar" )
```
