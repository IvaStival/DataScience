```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df_first = pd.read_csv("../hists/pred__Conv_2x2_3x3_3x3_Dense_64_64_Filter_4_4_8.csv")
df_sec   = pd.read_csv("../hists/pred__Conv_3x3_3x3_3x3_Dense_64_128_Filter_4_8_16.csv")
df_terc  = pd.read_csv("../hists/pred__Conv_5x5_3x3_3x3_Dense_128_256_Filter_8_16_32.csv")
df_quart = pd.read_csv("../hists/pred__Conv_5x5_5x5_5x5_Dense_256_256_Filter_16_32_64.csv")
df_quint = pd.read_csv("../hists/pred__Conv_5x5_8x8_4x4_Dense_256_512_Filter_32_32_64.csv")
```

```python
df_first.rename(columns={'Unnamed: 0':"signs"}, inplace=True)
df_sec.rename(columns={'Unnamed: 0':"signs"}, inplace=True)
df_terc.rename(columns={'Unnamed: 0':"signs"}, inplace=True)
df_quart.rename(columns={'Unnamed: 0':"signs"}, inplace=True)
df_quint.rename(columns={'Unnamed: 0':"signs"}, inplace=True)
```

```python
title_first = "Conv 2x2 3x3 3x3 Dense 64,64 Filter 4,4,8"
title_sec   = "Conv 3x3 3x3 3x3 Dense 64,128 Filter 4,8,16"
title_terc  = "Conv 5x5 3x3 3x3 Dense 128,256 Filter 8,16,32"
title_quart = "Conv 5x5 5x5 5x5 Dense 256,256 Filter 16,32,64"
title_quint = "Conv 5x5 8x8 4x4 Dense 256,512 Filter 32,32,64"
```

```python
df_first["signs"].size
```

```python
def plotPrecision(df, title, path="../plots/"):
    fig = plt.figure(figsize=(16, 10))
    b = sns.barplot(data=df, x="signs", y="precision")
    b.axes.set_title(title,fontsize=30)
    b.set_xlabel("Transit Signs",fontsize=20)
    b.set_ylabel("Precision",fontsize=20)
    b.tick_params(labelsize=15)
    plt.xticks(rotation=90)
    plt.tight_layout()
    
    name = title.replace(" ", "_").replace(",", "_")
    plt.savefig(f"{path}Precision_{name}.png", bbox_inches='tight')
```

```python
def plotRecall(df, title, path="../plots/"):
    fig = plt.figure(figsize=(16, 10))
    b = sns.barplot(data=df, x="signs", y="recall")
    b.axes.set_title(title,fontsize=30)
    b.set_xlabel("Transit Signs",fontsize=20)
    b.set_ylabel("Precision",fontsize=20)
    b.tick_params(labelsize=15)
    plt.xticks(rotation=90)
    plt.tight_layout()
    
#     name = title.replace(" ", "_").replace(",", "_")
#     plt.savefig(f"{path}Precision_{name}.png", bbox_inches='tight')
```

```python
plotPrecision(df_first, title_first) 
```

```python
plotPrecision(df_sec, title_sec) 
```

```python
plotPrecision(df_terc, title_terc) 
```

```python
plotPrecision(df_quart, title_quart) 
```

```python
plotPrecision(df_quint, title_quint) 
```

```python
plotRecall(df_first, title_first)
```

```python
plotRecall(df_quart, title_quart)
```
