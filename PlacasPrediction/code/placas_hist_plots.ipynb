```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df_first = pd.read_csv("../hists/Step01/hist__Conv_2x2_3x3_3x3_Dense_64_64_Filter_4_4_8.csv").iloc[: , 1:]
df_sec   = pd.read_csv("../hists/Step01/hist__Conv_3x3_3x3_3x3_Dense_64_128_Filter_4_8_16.csv").iloc[: , 1:]
df_terc  = pd.read_csv("../hists/Step01/hist__Conv_5x5_3x3_3x3_Dense_128_256_Filter_8_16_32.csv").iloc[: , 1:]
df_quart = pd.read_csv("../hists/Step01/hist__Conv_5x5_5x5_5x5_Dense_256_256_Filter_16_32_64.csv").iloc[: , 1:]
df_quint = pd.read_csv("../hists/Step01/hist__Conv_5x5_8x8_4x4_Dense_256_512_Filter_32_32_64.csv").iloc[: , 1:]
```

```python
epochs = pd.Series(range(0,df_first.shape[0]))
title_first = "Conv 2x2 3x3 3x3 Dense 64,64 Filter 4,4,8"
title_sec   = "Conv 3x3 3x3 3x3 Dense 64,128 Filter 4,8,16"
title_terc  = "Conv 5x5 3x3 3x3 Dense 128,256 Filter 8,16,32"
title_quart = "Conv 5x5 5x5 5x5 Dense 256,256 Filter 16,32,64"
title_quint = "Conv 5x5 8x8 4x4 Dense 256,512 Filter 32,32,64"
```

```python
def plotHists(df, epochs, y_value1, y_value2, title, label1, label2, x_label, y_label, color1="blue", color2="orange", path="../plots/"):
    fig, ax = plt.subplots(figsize=(14, 8))
#     ax.set(ylim=(0, 1))
    ax.plot(epochs, y_value1, data=df, color=color1, label=label1)
    ax.plot(epochs, y_value2, data=df, color=color2, label=label2)
    ax.set_title(title, fontsize=20)
    ax.set_xlabel(x_label, fontsize=16)
    ax.set_ylabel(y_label, fontsize=16)
    leg = ax.legend(prop={'size': 18})
    
    name = title.replace(" ", "_").replace(",", "_")
    plt.savefig(f"{path}Hist_{name}.png")
    plt.show()
```

```python
plotHists(df_first, epochs, "loss", "val_loss", title_first, "Train Loss", "Test Loss", "Epochs", "Loss")
```

```python
plotHists(df_first, epochs, "accuracy", "val_accuracy", title_first, "Train Acc", "Test Acc", "Epochs", "Accuracy")
```
